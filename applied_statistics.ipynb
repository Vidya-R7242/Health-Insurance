```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#lets load the insurance dataset and call it as ins_data & look at the head of the dataset
ins_data=pd.read_csv('insurance.csv')
ins_data.head()
```

```python
#lets take 5 point summery of the dataset.By looking at the below table we can infer that the
#'count' is same for every attribute; that means there are no missing values.
ins_data.describe(include='all')
```

```python
#lets check wheather there are any missing values. As the output below shows 
#'False' for all rows and columns; that means there are no missing values.
ins_data.isnull()
```

```python
#let's plot histogram for bmi, age & charges column.
#The distplot of bmi we can see that the data is distributed in normal fashion.
sns.distplot(ins_data['bmi'])
```

```python
#here the age is covering wide area that means the dataset consists of info about people 
#belonging to all age groups.
sns.distplot(ins_data['age'])
```

```python
#we can infer that the max charges is not normaly distibuted which is quite obvoius.since there
#are too many parameters affecting the output(charges)
sns.distplot(ins_data['charges'])
```

```python
#lets check how bmi, age & charges are related to one another.

sns.pairplot(ins_data[['age','bmi','charges']])
```

```python
#Skewness is a measure of the asymmetry of the probability distribution of a 
#real-valued random variable about its mean.
ins_data.skew()
```

```python
#a)Do charges of people who smoke differ significantly from the people who don't?
#from the below graph we can infer that the charges are much higher for smokers than
#for non-smokers. outliers are also more for non-smokers.

sns.boxplot(ins_data['smoker'],ins_data['charges'])
```

```python
#b)does bmi of males differ significantly from that of females?
#bmi for male and female varies almost in similat pattern.Few outliers are present for
# both attributes when plotted against bmi.

sns.boxplot(ins_data['sex'],ins_data['bmi'])
```

```python
#c)is the proportion of smokers significantly different in different genders?
#By looking at the numbers below in the chart we can say that the number of smokers in male is
#little higher than that of female.
pd.crosstab(ins_data['sex'],ins_data['smoker'])
```

```python
#this graph gives the picture of how many male & female have 0,1,2,3,4,5 children.
A=sns.countplot(x='children',hue='sex',data=ins_data)
```
